```python
import matplotlib.pyplot as plt
import torch
```

```python
bs = 10000
x = torch.randn(bs)
t = x.sigmoid()
plt_ts = torch.stack([x, t], dim=-1)

# bin plot
plt.hist(t, bins=100)
```

```python
rnd_norm = torch.randn((10000,))
sigma = rnd_norm.exp()
time_inv = sigma / (1 + sigma)
# time_inv = 2 / torch.pi * torch.atan(sigma)
plt.hist(1 - time_inv, bins=100)
```
