```python
t = pxt.create_table(
    "fundamentals.image_ops", {"source": pxt.Image}, if_exists="replace"
)
```

```python
url_prefix = (
    "https://github.com/pixeltable/pixeltable/raw/release/docs/resources/images"
)
images = ["000000000139.jpg", "000000000632.jpg", "000000000872.jpg"]
t.insert({"source": f"{url_prefix}/{image}"} for image in images)
```

```python
t.collect()
```

```python
t.add_computed_column(metadata=t.source.get_metadata())
t.collect()
```

```python
t.add_computed_column(rotated=t.source.rotate(10))
```

```python
t.collect()
```

```python
t.show()
```

```python
t.where(t.metadata["width"] >= 640).show()
```

```python
from pixeltable.iterators.image import TileIterator

t_v = pxt.create_view(
    "img_view",
    t,
    iterator=TileIterator.create(image=t.source, tile_size=(64, 64), overlap=(0, 0)),
)

t_v
```

```python
t_v.show()
```

```python
d = t_v.collect()
```

```python
ds = t_v.to_pytorch_dataset()
```
